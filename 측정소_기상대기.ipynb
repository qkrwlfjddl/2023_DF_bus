```python
import pandas as pd
import datetime
```

```python
air = pd.read_csv('/content/drive/MyDrive/DF/DF - 수소버스/2022년 일별평균대기오염도_2022.csv', encoding="cp949")
air.head()
```

```python
weather = pd.read_csv("/content/drive/MyDrive/DF/DF - 수소버스/일별기상.csv", encoding='cp949')
weather.head()
```

```python
format = '%Y-%m-%d'
air['측정일시'] = air['측정일시'].apply(lambda x: pd.to_datetime(str(x), format= format))
weather['day'] = weather['day'].apply(lambda x: pd.to_datetime(str(x), format= format))
```

```python
df = pd.merge(air, weather, left_on='측정일시', right_on='day')
df.drop(['day'],axis=1, inplace=True)
df.head()
```

```python
df = df.fillna(df.median())
df.isnull().sum()
```

```python
#변수선택한 독립변수들만 추출
col = ['측정일시', '측정소명', 'W_s','Dew','Press','See_1', '일산화탄소농도(ppm)','미세먼지농도(㎍/㎥)', '초미세먼지농도(㎍/㎥)']
df = df[col]
```

```python
def change_value(x):
  if x >= 18:
    return 1
  else:
    return 0
```

```python
df.iloc[:,8] = df.iloc[:,8].apply(change_value)
df['초미세먼지농도(㎍/㎥)'].value_counts()
```

# 모델링

```python
#train, test 분리
from sklearn.model_selection import train_test_split
feature = df.iloc[:,:8]
target = df['초미세먼지농도(㎍/㎥)']

X_train, X_test, y_train, y_test = train_test_split(feature, target, test_size=0.3, stratify=target, random_state=100)
```

```python
#oversampling
from imblearn.over_sampling import RandomOverSampler
X_train2, y_train2 = RandomOverSampler(random_state=0).fit_resample(X_train, y_train)
X_test2, y_test2 = RandomOverSampler(random_state=0).fit_resample(X_test, y_test)
```

```python
X_test2
```

```python
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
```

```python
# 종속변수가 이진 모델 (binomial) -> 로지스틱 모델 사용
from sklearn.linear_model import LogisticRegression

# 로지스틱 회귀 모델 생성 (L1 정규화 적용)
# max_iter 매개변수 설정하지 않으면 수렴 실패 -> max_iter 매개변수 늘리기
logit_model = LogisticRegression(penalty='l1', solver='liblinear', random_state=100, max_iter=1000)

# 모델 학습
logit_model.fit(X_train2.iloc[:,2:8], y_train2)

# 모델 예측
y_pred = logit_model.predict(X_test2.iloc[:,2:8])
```

```python
# 모델 성능 평가
from sklearn.metrics import roc_auc_score, roc_curve, accuracy_score, precision_score, recall_score, f1_score

auc = roc_auc_score(y_test2, y_pred)
accuracy = accuracy_score(y_test2, y_pred)
recall = recall_score(y_test2, y_pred)
precision = precision_score(y_test2, y_pred)
f1 = f1_score(y_test2, y_pred)

print("AUC:", auc)
print("Accuracy:", accuracy)
print("Recall:", recall)
print("Precision:", precision)
print("F1 스코어:", f1)
```

```python
X_test2['pred'] = y_pred
result = X_test2['측정소명'].loc[X_test2['pred']==1].value_counts().to_frame()
result.info()
```
